```python
# import library

from skimage import io
from skimage.color import rgb2gray
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import fftpack
import numpy as np
import cv2
```

```python
# load image

image = io.imread('moonlanding.png', as_gray=True)
```

```python
# shape image

image.shape, image.dtype
```

```python
f, ax = plt.subplots(figsize=(8,8))
ax.imshow(image, cmap = "gray")
ax.set_title('Original Noisy Image')
```

```python
# Return multidimensional discrete Fourier transform.

F = fftpack.fft2(image) 
F_magnitude = np.abs(F)   #absolute value of F
```

```python
# Show the results

def plot_spectrum(F):
    from matplotlib.colors import LogNorm
    plt.imshow(np.abs(F), norm=LogNorm(vmin=5))
    plt.colorbar()
```

```python
# A logarithmic Colormap

plt.figure(figsize=(8,8))
# plot_spectrum(F)
# plt.imshow(np.log(abs(F)), cmap = 'gray')
plt.imshow(np.log(abs(F)))
plt.title('Fourier transform')
```

```python
# Using shift to centering high frequency

Fs = fftpack.fftshift(F)
```

```python
plt.figure(figsize=(8,8))
# plot_spectrum(Fs)
#plt.imshow(np.log(abs(Fs)), cmap = 'gray')
plt.imshow(np.log(abs(Fs)))
plt.title('Shift Transformation Result')
```

for j in range(115, 126):
    for i in range (12, 22):

        Fs[i,j] = 0

    for i in range (298, 308):
        Fs[i,j] = 0

```python
# set to 0 according to the position of the noise

Fs[12:22, 115:125] = 0
Fs[37:47, 115:125] = 0
Fs[96:106, 115:125] = 0
Fs[216:226, 115:125] = 0
Fs[274:284, 115:125] = 0
Fs[298:308, 115:125] = 0
```

```python
plt.figure(figsize=(8,8))
# plt.imshow(np.log(abs(Fs)));
# plt.imshow(np.log(abs(Fs)), cmap = 'gray');
plt.imshow(np.log(abs(Fs)));
plt.title('Modified Fourier Transform');
```

```python
# Unshift

Fus = fftpack.ifftshift(Fs)
```

```python
# An Unshift logarithmic Colormap

plt.figure(figsize=(8,8))
# plot_spectrum(Fus)
# plt.imshow(np.log(abs(Fus)), cmap = 'gray')
plt.imshow(np.log(abs(Fus)))
plt.title('Unshift Spectrum')
```

```python
# Inverse FFT

im_new = fftpack.ifft2(Fus).real
```

```python
plt.figure(figsize=(8,8))
plt.imshow(im_new, plt.cm.gray)
plt.title('Reconstructed Image')
```
