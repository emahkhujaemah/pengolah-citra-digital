```python
from PIL import Image
```

```python
CITRA = Image.open('sekolah.jpg')
```

```python
ukuran_horizontal = CITRA.size[0]
ukuran_vertikal = CITRA.size[1]
```

```python
PIXEL = CITRA.load()
```

```python
for x in range(ukuran_horizontal):
    for y in range(ukuran_vertikal):
        R = 255 - PIXEL[x, y][0]
        G = 255 - PIXEL[x, y][1]
        B = 255 - PIXEL[x, y][2]
        PIXEL[x, y] = (R, G, B)
```

```python
CITRA.save('gambar_negatif.jpg')
```
