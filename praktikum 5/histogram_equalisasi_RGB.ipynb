```python
# Import library

import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Load image

image = cv2.imread('sekolah.jpg')
```

```python
plt.imshow(image)
```

```python
# Convert to RGB

img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
# Show the normal image

plt.imshow(img)
```

```python
# Show the histogram

hist,bins = np.histogram(img.flatten(),256,[0,256])
```

```python
# Visualization

plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.show()
```

```python
# equalize the histogram of the Y channel

img[:,:,0] = cv2.equalizeHist(img[:,:,0])
```

```python
# convert the YUV image back to RGB format

img_output = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
# Show the histogram

hist,bins = np.histogram(img_output.flatten(),256,[0,256])
plt.hist(img_output.flatten(),256,[0,256], color = 'g')
plt.xlim([0,256])
plt.show()
```

```python
# stacking both the images side-by-side orientation

res = np.hstack((image, img_output))
```

```python
# showing image input and output

cv2.imshow('image', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
