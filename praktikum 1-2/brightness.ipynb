```python
# import library

from PIL import Image
```

```python
# definisikan batas brightness (kecerahan)

def bright(intensitas):
    if intensitas < 0:
        return 0
    if intensitas > 255:
        return 255
    return intensitas
```

```python
# atur brightness

def atur_bright(nilai_pencerahan, nama_setelah_disave):
    CITRA = Image.open('sekolah.jpg')
    PIXEL = CITRA.load()
    ukuran_horizontal = CITRA.size[0]
    ukuran_vertikal = CITRA.size[1]

    for x in range(ukuran_horizontal):
        for y in range(ukuran_vertikal):
            R = bright(PIXEL[x, y][0] + nilai_pencerahan)
            G = bright(PIXEL[x, y][1] + nilai_pencerahan)
            B = bright(PIXEL[x, y][2] + nilai_pencerahan)
            PIXEL[x, y] = (R, G, B)

    CITRA.save(nama_setelah_disave)
```

```python
atur_bright(160, 'gambar_dicerahkan.jpg')
atur_bright(-150, 'gambar_digelapkan.jpg')
```
